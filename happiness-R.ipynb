```r
dh <- df  %>% group_by(Region) %>% summarise(count=n(),median_happiness=mean(H_Rank), max_life = max(Life_Exp), mean_generosity = mean(Generosity), mean_Family=mean(Family),dystopia=mean(Dystopia)) %>% arrange(desc(count)) %>% mutate(Region=factor(Region,Region))
dh
h <- ggplot(dh,aes(Region,n))
```

```r
h <- ggplot(dh,aes(x=Region,y=count, fill=Region))
h + geom_histogram(stat='identity')  + theme(axis.text.x = element_text(angle = 60, hjust = 1)) + scale_fill_brewer(palette = "RdBu") + ggtitle('Number of Countries per Region') +xlab('')
```

```r
dd <- df  %>% group_by(Region)%>% 
summarise(median_happiness=median(H_Rank), mean_happiness=mean(H_Rank),max_life = max(Life_Exp), mean_generosity = mean(Generosity),sd_generosity=sd(Generosity), mean_Family=mean(Family),dystopia=mean(Dystopia)) %>% 
arrange(max_life) %>% mutate(Region=factor(Region,Region))

de <- df  %>% group_by(Region)%>% 
summarise(count=n(),median_happiness=median(H_Rank), mean_happiness=mean(H_Rank),max_life = max(Life_Exp), mean_generosity = mean(Generosity),sd_generosity=sd(Generosity), mean_Family=mean(Family),dystopia=mean(Dystopia)) %>% 
arrange(mean_happiness) %>% mutate(Region=factor(Region,Region),inv_happiness = 1/mean_happiness)
dd
de
```

```r
d <- ggplot(dd,aes(Region,max_life,fill=median_happiness))
d1 <- ggplot(dd,aes(Region,max_life,fill=mean_happiness))
d1 + geom_bar(stat='identity') + theme(axis.text.x = element_text(angle = 60, hjust = 1)) + xlab('') + ylab('Max Life Expectancy') + ggtitle('Max Life Expectancy by Region')

d + geom_bar(stat='identity') + theme(axis.text.x = element_text(angle = 60, hjust = 1)) + xlab('') + ylab('Max Life Expectancy') + ggtitle('Max Life Expectancy by Region')

e <- ggplot(de,aes(Region,inv_happiness,fill=max_life))
e + geom_bar(stat='identity') + theme(axis.text.x = element_text(angle = 60, hjust = 1)) + xlab('') + ylab('Average Happiness Rank') + ggtitle('Average Happiness by Region')


j <- ggplot(df, aes(x=reorder(Region,H_Score,mean),y=H_Score,fill=Region),alpha=.8)
j + geom_boxplot() + theme(axis.text.x = element_text(angle = 60, hjust = 1)) + scale_fill_brewer(palette = "RdBu")
```

```r
normalize <- function(x) {
num <- x - min(x)
denom <- max(x) - min(x)
return (num/denom)
}
```

```r
df_norm <- as.data.frame(lapply(df[3:12], normalize))

# Summarize `iris_norm`
summary(df_norm)

df_norm_ <- cbind(df$Region,df_norm)
colnames(df_norm_)[1] = 'Region'
ggplot(df,aes(Life_Exp,H_Score,color=Region),alpha=.1) + geom_point() + geom_smooth(method=lm, se=FALSE) + theme_minimal()+ scale_fill_brewer(palette = 'RdBu') + ylab('Happines Score') + xlab('Life Expectancy') +
ggtitle('Happiness V Life Expectancy ~ 2015') + facet_wrap(~Region)colnames(df_norm_)
```
